```python
import pandas as pd

# Load Dataset
df = pd.read_excel("Online_Retail_Small.xlsx")

# Clean Data
df = df.dropna(subset=['CustomerID'])
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Create Total Amount
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']

# Calculate Total Spending Per Customer
customer_df = df.groupby('CustomerID')['TotalAmount'].sum().reset_index()

# Create Customer Segments
def segment_customer(amount):
    if amount >= 1000:
        return "High Value Customer"
    elif amount >= 500:
        return "Medium Value Customer"
    else:
        return "Low Value Customer"

customer_df['Segment'] = customer_df['TotalAmount'].apply(segment_customer)

# Save as CSV
customer_df.to_csv("customer_segments.csv", index=False)

print("Customer Segmentation Completed Successfully!")
print("CSV file created successfully!")

# Search Customer
customer_id = float(input("\nEnter Customer ID: "))

customer = customer_df[customer_df['CustomerID'] == customer_id]

if not customer.empty:
    print("\n===== CUSTOMER DETAILS =====")
    print("Customer ID :", int(customer.iloc[0]['CustomerID']))
    print("Total Spending :", round(customer.iloc[0]['TotalAmount'], 2))
    print("Segment :", customer.iloc[0]['Segment'])
else:
    print("Customer ID not found.")
```
